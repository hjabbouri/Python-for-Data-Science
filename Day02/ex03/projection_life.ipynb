```python
import matplotlib.pyplot as plt
from load_csv import load

# TODO: add __doc__


def main():
    loaded_income_pp = load('../income_per_person_gdppercapita_ppp_inflation_adjusted.csv')
    loaded_life_expectancy = load('../life_expectancy_years.csv')
    income_pp = loaded_income_pp['1900']
    life_expectancy = loaded_life_expectancy['1900']
    # life_expectancy.fillna(0, inplace=True)
    plt.scatter(income_pp, life_expectancy, label='Life Expectancy in 1900')
    plt.xlabel('Gross domestic product')
    plt.xscale('log')
    plt.xlim(300, 11000)
    plt.xticks([300, 1000, 10000], ['300', '1K', '10K'])
    plt.ylabel('Life Expectancy')
    plt.legend(loc='lower right')
    plt.title('1900')
    plt.show()


if __name__ == '__main__':
    main()
```
